```python
Scagnostics = ["Outlying","Skewed","Clumpy","Sparse","Striated","Convex","Skinny","Stringy","Monotonic"]
# Variables = list(data['compute-1-1'].keys())
# Variables = ['arrTemperature', 'arrCPU_load', 'arrMemory_usage', 'arrFans_health', 'arrPower_usage']
Variables = ['arrTemperature', 'arrMemory_usage', 'arrFans_health', 'arrPower_usage']
# Variables = ['arrTemperature', 'arrCPU_load', 'arrMemory_usage', 'arrFans_health']
# Countries = [x for x in list(data.keys()) if x != 'timespan']
Countries = [x for x in list(data.keys()) if x != 'timespan'] # for dataset influxdbThus21Mar_1400_1630.json also filter out compute-1-26 since there is no data.
```

```python
lens = []
for c in Countries:
    for v in Variables:
        if v not in data[c]:
            data[c][v] = [[None, None, None]]
        if(len(data[c][v])!=timeSteps):
            #Then repeat the data
            for i in range(len(data[c][v]), timeSteps):
                data[c][v].append([None]*len(data[c][v][0]))
        lens.append(len(data[c][v]))
```

```python
CountriesData = {}

for c in Countries:
    items = []
    for time_i in range(timeSteps):
        item = {}
        var_counter = 0
        for i, v in enumerate(Variables):
            if(v in data[c].keys() and len(data[c][v])!=0):
                for k in range(len(data[c][v][time_i])):
                    if(k <= 3):
                        if ((v=='arrCPU_load' or v=='arrPower_usage' or v=='arrMemory_usage') and k>0):#skipping the None fields
                            continue
                        item['v'+str(var_counter)] = data[c][v][time_i][k]
                        var_counter += 1
            else:
                item['v'+str(var_counter)] = None
                var_counter += 1
            item['year'] = time_i
            item['Outlying'] = 0
        items.append(item)
    CountriesData[c] = items    
```

```python
CountriesDataVariables = [v for v in CountriesData[Countries[0]][0].keys() if v not in ['year', 'Outlying']]
```

```python
RawYearsData = []
for time_i in range(timeSteps):
    items = []
    for c in Countries:
        item = {}
        for i, v in enumerate(CountriesDataVariables):
            if v in CountriesData[c][time_i].keys():
                item['v'+str(i)] = CountriesData[c][time_i][v]
            else:
#                 item['v'+str(i)] = None
                item['v'+str(i)] = 0
        items.append(item)
    RawYearsData.append(items)
```

```python
RawYearsData[0]
```

```python
MinMaxYearsData = []
#Initialize
for time_i in range(timeSteps):
    item = {}
    for i, v in enumerate(CountriesDataVariables):
        item['v'+str(i)] = {'max': float('-inf'), 'min':float('inf')}
    MinMaxYearsData.append(item)
#Calculate
for time_i in range(timeSteps):#year
    for c_i, c in enumerate(Countries):#country
        for v_i, v in enumerate(CountriesDataVariables):#variable
            val = RawYearsData[time_i][c_i]['v'+str(v_i)]
            if(val != None):
                if(MinMaxYearsData[time_i]['v'+str(v_i)]['min'] > val):
                    MinMaxYearsData[time_i]['v'+str(v_i)]['min'] = val
                if(MinMaxYearsData[time_i]['v'+str(v_i)]['max'] < val):
                    MinMaxYearsData[time_i]['v'+str(v_i)]['max'] = val
```

```python
NormalizedYearsData = []
for time_i in range(timeSteps):#year
    yearData = {'Scagnostics0': [0 for i in range(len(Scagnostics))]}
    for v_i, v in enumerate(CountriesDataVariables):#variable
        yearData['s'+str(v_i)] = []
        for c_i, c in enumerate(Countries):#country
            mm = MinMaxYearsData[time_i]['v'+str(v_i)]
            val = RawYearsData[time_i][c_i]['v'+str(v_i)]
            if(val!=None):
                normalized_val = (val-mm['min'])/(mm['max'] - mm['min']) if mm['max'] != mm['min'] else 0.5
                yearData['s'+str(v_i)].append(normalized_val)
            else:
#                 yearData['s'+str(v_i)].append(float('nan'))
                yearData['s'+str(v_i)].append(None)
    NormalizedYearsData.append(yearData)
```
